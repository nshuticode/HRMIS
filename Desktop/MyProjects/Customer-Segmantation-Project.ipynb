# Customer Segmentation

```python
data['DayOfWeek'] = pd.to_datetime(data['Date']).dt.dayofweek   
```

```python
#total purchase amount
data["TotalAmount"] = data["Quantity"] * data["UnitPrice"]
```

```python
print(data)
```

```python
# Recency (R)
recency_data = data.groupby('CustomerID')['Date'].max()
recency_data = pd.to_datetime('2023-08-24')

# Frequency (F)
frequency_data = data.groupby('CustomerID')['InvoiceNo'].nunique()

# Monetary (M)
monetary_data = data.groupby('CustomerID')['TotalAmount'].sum()

# Average Purchase Quantity
avg_quantity_data = data.groupby('CustomerID')['Quantity'].mean()

# Average Unit Price
avg_unit_price_data = data.groupby('CustomerID')['UnitPrice'].mean()

# Country Diversity
country_diversity_data = data.groupby('CustomerID')['Country'].nunique()

# Calculate list of unique countries for each customer
unique_countries_data = data.groupby('CustomerID')['Country'].unique()

# Shopping Hour Preference
shopping_hour_data = data.groupby('CustomerID')['Hour'].mean()

# Create the RFM DataFrame
rfm_data = pd.DataFrame({
    'Recency': recency_data,
    'Frequency': frequency_data,
    'Monetary': monetary_data
})

# Create the additional_features DataFrame
additional_features = pd.DataFrame({
    'CustomerID': avg_quantity_data.index,
    'avg_quantity_data': avg_quantity_data.values,
    'avg_unit_price_data': avg_unit_price_data.values,
    'country_diversity_data': country_diversity_data.values,
    'shopping_hour_data': shopping_hour_data.values
})

# Merge the RFM features with the additional features based on 'CustomerID' column
customer_data = rfm_data.merge(additional_features, on='CustomerID')


print(customer_data.head())
```

# Visualization and Segment Analysis

```python
# Convert Recency to numerical representation of days since the minimum date
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd

# Convert 'Recency' to datetime format if it's not already
customer_data['Recency'] = pd.to_datetime(customer_data['Recency'])

# Calculate the numerical representation of days since the minimum date
min_date = customer_data['Recency'].min()
customer_data['Recency'] = (min_date - customer_data['Recency']).dt.days

# Select features for clustering
X = customer_data[['Recency', 'Frequency', 'Monetary', 'avg_quantity_data', 'avg_unit_price_data', 'country_diversity_data', 'shopping_hour_data']]

silhouette_scores = []
wcss = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, n_init=8, random_state=42)  # Set n_init to 10 explicitly
    kmeans.fit(X)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(X, labels))
    wcss.append(kmeans.inertia_)

# Create a 1x2 grid for plots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Silhouette Scores
axes[0].plot(range(2, 11), silhouette_scores, marker='o')
axes[0].set_xlabel('Number of Clusters')
axes[0].set_ylabel('Silhouette Score')
axes[0].set_title('Silhouette Score vs Number of Clusters')

# Plot WCSS
axes[1].plot(range(2, 11), wcss, marker='o', color='orange')
axes[1].set_xlabel('Number of Clusters')
axes[1].set_ylabel('WCSS')
axes[1].set_title('WCSS vs Number of Clusters')

# Add title to the whole figure
fig.suptitle('Cluster Analysis', fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plots
plt.show()

```

```python
#mean values of each feature within each cluster.
# Assuming you have already performed the clustering and selected the optimal number of clusters
optimal_num_clusters = 5  # Replace this with the actual number of clusters

# Fit the KMeans model with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_num_clusters, n_init=10, random_state=42)
kmeans.fit(X)

# Assign each data point to a cluster
customer_data['Cluster'] = kmeans.labels_

# Calculate the mean values of each feature within each cluster
cluster_means = customer_data.groupby('Cluster').mean()

# Display the mean values for each cluster
print(cluster_means)
```

```python
# Apply t-SNE to reduce the dimensionality of the data and visualize clusters
tsne = TSNE(n_components=2, perplexity=30, n_iter=300, random_state=42)
tsne_result = tsne.fit_transform(X)

# Add cluster labels to the t-SNE results
tsne_df = pd.DataFrame(data=tsne_result, columns=['t-SNE1', 't-SNE2'])
tsne_df['Cluster'] = kmeans.labels_

# Create a scatter plot of the t-SNE results colored by cluster
plt.figure(figsize=(10, 6))
plt.scatter(tsne_df['t-SNE1'], tsne_df['t-SNE2'], c=tsne_df['Cluster'], cmap='rainbow')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Visualization of Clusters')
plt.colorbar(label='Cluster')
plt.show()
```
